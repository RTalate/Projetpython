```python
from selenium.common.exceptions import NoSuchElementException
```

```python
lt_url = []
for i, j in dic_id_team.items():
  
  
  for name_team,id_team in dic_id_team.items():
    url = model_url.replace('id_team_replace',id_team).replace('years_replace',i).replace('NAMETEAM_replace',name_team).replace('id_year_replace',j)
    wd.get(url)
  
    try:
      wd.find_element_by_xpath('//*[@id="matchlogs_for"]/tbody')
      l = str(wd.find_element_by_xpath('//*[@id="matchlogs_for"]/tbody'))

      lt =  []
      for i in re.findall(r'href="(.*?)">',l):
        if 'Premier-League' in i:
          lt.append(i)
      lt = list(set(list(lt)))
  
      for i in lt:
        lt_url.append('https://fbref.com'+i)
      pass
    except NoSuchElementException:
      pass
```
